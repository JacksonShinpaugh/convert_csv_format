```python
import os
import ast
import pandas as pd
```

```python
df = pd.read_csv('pf_detections.csv')
df.head()
```

```python
data = []



for ind, row in df.iterrows():
    video_name = row['file_name']
    det_per_sec = ast.literal_eval(row['detections_per_second'])
    
    for time, detections in det_per_sec.items():

        for label, value in detections.items():
            data.append({
                "video_name": video_name,
                "time_stamp": time,
                "label": label,
                "value": value
            })

df = pd.DataFrame(data)
df.to_csv('pf_detections_reformatted.csv') 
```
