```python
import os
import json
import pandas as pd
```

```python
df = pd.read_csv('pf_detections2.csv')
df.head()
```

```python
data = []

for ind, row in df.iterrows():
    video_name = row['file_name']
    dps = row["detections_per_second"].replace("'", '"')
    json_file = json.loads(dps)
    
    for time, detections in json_file.items():

        for label, value in detections.items():
            data.append({
                "video_name": video_name,
                "time_stamp": time,
                "label": label,
                "value": value
            })

df = pd.DataFrame(data)
df.to_csv('pf_detections_reformatted.csv') 
```
