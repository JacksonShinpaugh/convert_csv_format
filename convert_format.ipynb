```python
import os
import ast
import pandas as pd
```

```python
df = pd.read_csv('pf_detections.csv')
df.head()
```

```python
data = []

def extract_and_adjust_df(row):

    video_name = row['file_name']
    det_per_sec = ast.literal_eval(row['detections_per_second'])

    #goes through a dictionary inside each row of the above column
    for time, detections in det_per_sec.items():

        # goes throug heach element of the dictionary
        for label, value in detections.items():
            data.append({
                "video_name": video_name,
                "time_stamp": time,
                "label": label,
                "value": value
            }) 
    return


df.apply(extract_and_adjust_df, axis=1)
df = pd.DataFrame(data)
df.to_csv('pf_detections_reformatted.csv') 
```
